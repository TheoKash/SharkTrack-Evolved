```python
import pandas as pd
import os
import shutil

path = '/vol/biomedic3/bglocker/ugproj2324/fv220/datasets/annotations/mwitt/AXA_NOV23_v3'
new_path = '/vol/biomedic3/bglocker/ugproj2324/fv220/datasets/annotations/mwitt/AXA_NOV23_v3_non_empty_2'
empty = 0
non_empty = 0
folders = []
folders_empty = []

for subf in os.listdir(path):
  os.makedirs(os.path.join(new_path, subf), exist_ok=True)
  p = os.path.join(path, subf)
  annos = None
  for csv in os.listdir(p):
    annos = True
    # check if empty
    df = pd.read_csv(os.path.join(p, csv), skiprows=1)
    if df.empty:
      empty += 1
    else:
      non_empty += 1
      # copy to new path
      # shutil.copy(os.path.join(p, csv), os.path.join(new_path, subf, csv))
  
  if annos:
    folders.append(subf)
  else: 
    folders_empty.append(subf)
```

```python
# Split all the non-empty annotations in 7 folds. 
# for each fold, for each directory, print the annotations that are in the fold
# if possible, have annotations in the same directory in the same fold

print(f'Empty: {empty}, Non-empty: {non_empty}')
print(f'Empty: {empty/(empty+non_empty)*100:.2f}%, Non-empty: {non_empty/(empty+non_empty)*100:.2f}%')

# 7 folds
folds = [[] for _ in range(7)]
max_fold_size = non_empty // 7
i = 0
for subf in os.listdir(path):
  p = os.path.join(path, subf)
  for csv in os.listdir(p):
    # check if empty
    df = pd.read_csv(os.path.join(p, csv), skiprows=1)
    if df.empty:
      continue
    else:
      if len(folds[i]) > max_fold_size:
        if i == 6:
          print(f'max fold size, adding {csv} to fold {i}')
        else:
          i += 1
      folds[i].append(os.path.join(p, csv))
```

```python
import os
import shutil

new_path = '/vol/biomedic3/bglocker/ugproj2324/fv220/datasets/annotations/mwitt/AXA_NOV23_v3_non_empty'
os.makedirs(new_path, exist_ok=True)

for i, fold in enumerate(folds):
  fold_path = os.path.join(new_path, f'fold_{i}')
  os.makedirs(fold_path, exist_ok=True)
  
  for csv in fold:
    parent_dir = os.path.dirname(csv)
    parent_dir = os.path.basename(parent_dir)
    new_csv_path = os.path.join(fold_path, parent_dir)
    os.makedirs(new_csv_path, exist_ok=True)
    shutil.copy(csv, new_csv_path)
```

```python
import os
import pandas as pd

annotation_folder = '/vol/biomedic3/bglocker/ugproj2324/fv220/datasets/annotations/mwitt/AXA_APR23_v2/AXA_2023-0011704'

# Find MaxN: max number of simultaneous annotations in a video.
# Simultanous annotations are annotations at the same column: "3: Unique Frame Identifier"

maxN = 0
maxN_video = None
maxN_frame = None
for csv in os.listdir(annotation_folder):
  df = pd.read_csv(os.path.join(annotation_folder, csv), skiprows=lambda x: x in [1])
  # curr maxN = max number of rows with the same "3: Unique Frame Identifier"
  curr_maxN = df['3: Unique Frame Identifier'].value_counts().max()
  if curr_maxN > maxN:
    maxN = curr_maxN
    maxN_video = csv
    maxN_frame = df['3: Unique Frame Identifier'].value_counts().idxmax()

print(maxN, maxN_video, maxN_frame)
```
